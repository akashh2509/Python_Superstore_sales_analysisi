```python
# Filter data for the 'Tables' sub-category
tables_data = data[data['Sub-Category'] == 'Tables']

# Summarize the profit for tables
tables_profit = tables_data['Profit'].sum()
print(f"Total Profit for Tables: {tables_profit}")
```

### Finding the Reason for Losses in the Tables Sub Category

```python
# Plot profit distribution for tables to verify it is negative
tables_data['Profit'].plot(kind='hist', bins=30, color='red')
plt.title("Profit Distribution for Tables")
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.show()
# Now we find the correlation bw features and the feature which has most correlation with profits would most probably be the reason for losses.
# Scatter plot for Discount vs Profit
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Discount', y='Profit', data=tables_data, palette='coolwarm')
plt.title("Discount vs Profit for Tables")
plt.xlabel("Discount (in %)")
plt.ylabel("Profit")
plt.show()

# Scatter plot for Shipping Cost vs Profit
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Shipping Cost', y='Profit', data=tables_data, palette='coolwarm')
plt.title("Shipping Cost vs Profit for Tables")
plt.xlabel("Shipping Cost")
plt.ylabel("Profit")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Discount', y='Profit', size='Shipping Cost', data=tables_data, palette='coolwarm', sizes=(20, 200))
plt.title("Discount and Shipping Cost vs Profit for Tables")
plt.xlabel("Discount (in %)")
plt.ylabel("Profit")
plt.show()
```

### Using the visualizations we are able to draw the following insights -

#### Discount Distribution:

If the histogram shows high discounts frequently, this could be a major factor in the negative profit.
The scatter plot will indicate how discounts correlate with profit. If higher discounts often correspond to negative profits, it confirms the issue.

#### Shipping Cost Distribution:

The histogram will show if shipping costs are generally high.
The scatter plot will show the relationship between shipping costs and profit. High shipping costs corresponding to negative profits indicate another problem area.

#### Combined Factors:

The combined scatter plot will show how both high discounts and high shipping costs together affect profit. This can help in understanding if one or both factors are contributing significantly.
